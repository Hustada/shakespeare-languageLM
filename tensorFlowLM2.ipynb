```python
import csv
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Read in the .txt file as a Pandas DataFrame
lines = []
with open('shakespeare.txt') as f:
    for line in f:
        lines.append(line.strip())
df = pd.DataFrame(lines, columns=['text'])

# Split the data into a test set and a training set
X_train, X_test, y_train, y_test = train_test_split(df, df['text'], test_size=0.2)

# Split the training data into a validation set and a training set
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2)

# Save the data to separate files
X_train.to_csv('train.csv', index=False)
X_val.to_csv('validation.csv', index=False)
X_test.to_csv('test.csv', index=False)

# Read the train.csv file into a pandas DataFrame
train_df = pd.read_csv('train.csv')

# Print the first 5 rows of the DataFrame
print(train_df.head())

# Load the datasets
train_df = pd.read_csv('train.csv')
val_df = pd.read_csv('validation.csv')
test_df = pd.read_csv('test.csv')

# Split the data into input and target variables
X_train = train_df.drop('text', axis=1)
y_train = train_df['text']
X_val = val_df.drop('text', axis=1)
y_val = val_df['text']
X_test = test_df.drop('text', axis=1)
y_test = test_df['text']

# Convert the data to NumPy arrays
X_train = np.array(X_train)
y_train = np.array(y_train)
X_val = np.array(X_val)
y_val = np.array(y_val)
X_test = np.array(X_test)
y_test = np.array(y_test)

# Create a LabelEncoder object
le = LabelEncoder()

# Fit the LabelEncoder on the target data
le.fit(np.concatenate((y_train, y_val, y_test)))

# Transform the target data to numeric
y_train_numeric = le.transform(y_train)
y_val_numeric = le.transform(y_val)
y_test_numeric = le.transform(y_test)

# Define the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train the model
history = model.fit(X_train, y_train_numeric, epochs=100, validation_data=(X_val, y_val_numeric))

# # Split the data into input and target variables
# X_train = train_df.drop('text', axis=1)
# y_train = train_df['text']
# X_val = val_df.drop('text', axis=1)
# y_val = val_df['text']
# X_test = test_df.drop('text', axis=1)
# y_test = test_df['text']

# # Convert the data to NumPy arrays
# X_train = np.array(X_train)
# y_train = np.array(y_train)
# X_val = np.array(X_val)
# y_val = np.array(y_val)
# X_test = np.array(X_test)
# y_test = np.array(y_test)

# # Create a LabelEncoder object
# le = LabelEncoder()

# # Fit the LabelEncoder on the target data
# le.fit(np.concatenate((y_train, y_val, y_test)))

# # Transform the target data to numeric
# y_train_numeric = le.transform(y_train)
# y_val_numeric = le.transform(y_val)
# y_test_numeric = le.transform(y_test)

# # Define the model
# model = tf.keras.Sequential([
#     tf.keras.layers.Dense(32, input_shape=(X_train.shape[1],)),
#     tf.keras.layers.Dense(32, activation='relu'),
#     tf.keras.layers.Dense(1)
# ])

# # Compile the model
# model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# # Train the model
# history = model.fit(X_train, y_train_numeric, epochs=30, validation_data=(X_val, y_val_numeric))

# # Plot the training and validation loss
# plt.plot(history.history['loss'], label='Training loss')
# plt.plot(history.history['val_loss'], label='Validation loss')
# plt.legend()
# plt.show()

# # Evaluate the model on the test data
# test_loss, test_acc = model.evaluate(X_test, y_test_numeric)
# print('Test loss:', test_loss)
# print('Test accuracy:', test_acc)

```
